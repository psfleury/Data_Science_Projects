# Panorama do COVID-19 no Brasil

A COVID-19 é uma doença infecciosa causada pelo novo coronavírus (SARS-CoV-2) e tem como principais sintomas febre, cansaço e tosse seca. Alguns pacientes podem apresentar dores, congestão nasal, dor de cabeça, conjuntivite, dor de garganta, diarreia, perda de paladar ou olfato, erupção cutânea na pele ou descoloração dos dedos das mãos ou dos pés. Esses sintomas geralmente são leves e começam gradualmente. Algumas pessoas são infectadas, mas apresentam apenas sintomas muito leves.

A maioria das pessoas (cerca de 80%) se recupera da doença sem precisar de tratamento hospitalar. Uma em cada seis pessoas infectadas por COVID-19 fica gravemente doente e desenvolve dificuldade de respirar. As pessoas idosas e as que têm outras condições de saúde como pressão alta, problemas cardíacos e do pulmão, diabetes ou câncer, têm maior risco de ficarem gravemente doentes. No entanto, qualquer pessoa pode pegar a COVID-19 e ficar gravemente doente.

<center><img alt="Colaboratory logo" width="70%" src="https://raw.githubusercontent.com/psfleury/Data_Science_Projects/main/2%20-%20Covid19/virus-da-covid.jpg"></center>

Todos os vírus, incluindo o SARS-CoV-2, mudam com o tempo. Grande parte das mudanças tem pouco a nenhum impacto nas propriedades do vírus. Entretanto,de acordo com a Organização Mundial da Saúde (OMS), com a transmissão indiscriminada do vírus em diversas partes do mundo, não é raro que surjam mutações que afetem propriedades do vírus relevantes para a humanidade, tais como:
* Taxa de transmissibilidade;
* Severidade da infecção decorrente;
* Performance de vacinas;
* Adequação de medicamentos terapeuticos;
* Ferramentas de diagnóstico;
* Outras medidas de saúde pública e sociais.

As variantes do SARS-CoV-2 que caracterizam um aumento de risco à saúde pública global passaram a ser classificadas em duas categorias pela OMS: 

i) **Variantes de Interesse (*Variants of Interest - VOI*)**, as quais contêm características que afetam a transmissibilidade, severidade, escapamento à imunidade adquirida, ao diagnóstico **ou** a tratamentos terapeuticos e que estejam causando transmissão comunitária ou múltiplos *clusters* de Covid-19, em diversos países, com aumento temporal de sua prevalência;

ii) **Variantes Preocupantes (*Variants of Concern - VOC*)**, aquelas que se adequam à definição de Variante de Interesse (VOI) e, por meio de uma análise comparativa, tem demonstrado estar associada com **uma ou mais** das seguintes mudanças em termos de saúde pública global: a) aumento de transmissibilidade ou mudanças negativas na epidemiologia do Covid-19; ou b) Aumento de carga viral ou mudança na apresentação clínica da doença; ou c) Decréscimo na efetividade das medidas de saúde pública ou sociais ou dos diagnósticos disponíveis, das vacinas e procedimentos terapeuticos.

As Variantes Preocupantes (VOC), até 17/08/2021, são quatro, classificadas da seguinte forma:

* Alpha (B.1.1.7), surgida no Reino Unido em setembro/2020 e designada como VOC em 18/12/2020;

* Beta (B.1.351), surgida na África do Sul em maio/2020 e designada como VOC em 18/12/2020;

* Gamma (P.1), surgida no Brasil em novembro/2020 e designada como VOC em 11/01/2021; 

* Delta (B.1.617.2), surgida na Índia em outubro/2020 e designada como VOC em 11/05/2021.

Desde o fim de 2020, pelo avanço de tais Variantes Preocupantes (VOC),  embora tenham ocorridos avanços sem precedentes no desenvolvimento de vacinas contra a Covid-19, mesmo países com alta taxa de imunização vacinal continuem com dificuldade de conter o avanço da doença, como é exemplo o Estados Unidos.

### Dicionário de Variáveis

**Nome do atributo - Descrição**
* iso_code - Código de país ISO 3166, com 3 letras;
* continent - Continente;
* location - Localidade geográfica;
* date - Data da observação;
* total_cases - Quantidade acumulada de casos;
* new_cases - Quantidade de novos casos;
* new_cases_smoothed - Média móvel de novos casos em 7 dias;
* total_deaths - Quantidade acumulada de mortes;
* new_deaths - Quantidade de novas mortes;
* new_deaths_smoothed - Média móvel de mortes em 7 dias;
* total_cases_per_million - Quantidade acumulada de casos por milhão de pessoas;
* new_cases_per_million - Quantidade de novos casos por milhão de pessoas;
* new_cases_smoothed_per_million - Média móvel de novos casos em 7 dias por milhão de pessoas;
* total_deaths_per_million - Quantidade acumulada de mortes por milhão de pessoas;
* new_deaths_per_million - Quantidade de novas mortes por milhão de pessoas;
* new_deaths_smoothed_per_million - Média móvel de novas mortes em 7 dias por milhão de pessoas;
* reproduction_rate - Taxa de reprodução;
* icu_patients - Pacientes em UTI no dia em questão;
* icu_patients_per_million - Pacientes em UTI no dia em questão por milhão de pessoas;
* hosp_patients - Pacientes hospitalizados no dia em questão;
* hosp_patients_per_million - Pacientes hospitalizados no dia em questão por milhão de pessoas;
* weekly_icu_admissions - Admissões em UTI na semana;
* weekly_icu_admissions_per_million - Admissões em UTI na semana por milhão;
* weekly_hosp_admissions - Admissões em hospital na semana;
* weekly_hosp_admissions_per_million - Admissões em hospital na semana por milhão;
* total_tests - Quantidade acumulada de testes;
* new_tests - Quantidade de novos testes;
* total_tests_per_thousand - Quantidade acumulada de testes por mil pessoas;
* new_tests_per_thousand - Quantidade de novos testes por mil pessoas;
* new_tests_smoothed - Média móvel de 7 dias para novos testes;
* new_tests_smoothed_per_thousand - Média móvel de 7 dias para novos testes por mil pessoas;
* positive_rate - Média móvel de 7 dias do percentual de testes que deram positivo;
* tests_per_case - Média móvel de 7 dias de testes conduzidos por número de positivos identificados (inverso de 'positive_rate');
* tests_units - Unidades usadas pela localidade para reportar seus dados de teste;
* total_vaccinations - Quantidade total de vacinas administradas;
* people_vaccinated - Quantidade total de pessoas que receberam ao menos uma dose de vacina;
* people_fully_vaccinated - Quantidade total de pessoas que estão completamente vacinadas segundo o protocolo;
* total_boosters - Número total de doses extras ao protocolo administradas;
* new_vaccinations - Número de novas vacinas administradas;
* new_vaccinations_smoothed - Média móvel de 7 dias de novas vacinas administradas;
* total_vaccinations_per_hundred - Total de vacinas administradas por 100 pessoas;
* people_vaccinated_per_hundred - Total de pessoas que receberam ao menos uma dose por 100 pessoas;
* people_fully_vaccinated_per_hundred - Total de pessoas completamente vacinadas segundo o protocolo por 100 pessoas;
* total_boosters_per_hundred - Número de doses extras administradas por 100 pessoas;
* new_vaccinations_smoothed_per_million - Média móvel de 7 dias de novas doses administradas por milhão de pessoas;
* stringency_index - Índice de rigor da resposta governamental, composto por nove indicadores, incluindo fechamento de escolas, de ambientes de trabalho, banimento de viagens, etc. Tal índice encontra-se entre 0 e 100.;
* population - População do país em 2020;
* population_density - Densidade populacional;
* median_age - Idade mediana da população em 2020;
* aged_65_older - Percentual da população com 65 anos ou mais;
* aged_70_older - Percentual da população com 70 anos ou mais;
* gdp_per_capita - PIB per capita do ano mais recente disponível;
* extreme_poverty - Percentual da população vivendo em pobreza extrema;
* cardiovasc_death_rate - Taxa de mortes por problema cardiovascular por cem mil pessoas;
* diabetes_prevalence - Prevalência de diabetes na população (% da população com idade entre 20 e 79 anos);
* female_smokers - Percentual das mulheres que fumavam no ano mais recente;
* male_smokers - Percentual dos homens que fumavam no ano mais recente;
* handwashing_facilities - Percentual da população com acesso a ambientes para lavagem de mão;
* hospital_beds_per_thousand - Número de camas de hospital por 1000 pessoas, do ano mais recente;
* life_expectancy - Expectativa de vida no nascimento em 2019;
* human_development_index - Índice composto que mede atingimento em três dimensões básicas do desenvolvimento humano (Vida longa e saudável, conhecimento e decente padrão de vida), de acordo com valores de 2019.;
* excess_mortality - P-valor de mortalidade em excesso, para todas as idades.
